```python
def outlier_limit(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    iqr
    upper= q3+1.5*iqr
    lower= q1-1.5*iqr
    return upper,lower
for column in df_norm.columns:
    UL, LL = outlier_limit(df_norm[column])
    df_norm[column] = np.where((df_norm[column] > UL) | (df_norm[column] < LL), np.nan, df_norm[column])
```
